```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Dataset_Day15.csv')
df.head(20)
```

```python
# 1.	Read the data and remove all null or empty values.
from csv import reader

groceries =[]
with open('Dataset_Day15.csv','r') as csvfile:
    read = reader(csvfile)
    for row in read:
        row_new = [x for x in row if x!='']
        groceries.append(row_new)
```

```python
groceries[0:5]
```

```python
# 2.	Transform the data using TransactonEncoder to perform Market Basket Analysis.
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()

transactions = encoder.fit(groceries).transform(groceries)
transactions
```

```python
itemsets = pd.DataFrame(transactions, columns = encoder.columns_)

display(itemsets.head())
display(itemsets.info())
```

```python
# 3.	 Use min_support = 0.02 to find frequent itemsets.

from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(itemsets, min_support = 0.02, use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support', ascending = False)
```

```python
# 4.	Use the frequent itemsets to create association rules (take min_threshold as 15%) and evaluate the rules for the following metrics

from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.15)
rules
```

```python
# a.	Find top 5 antecedent -> consequent rules based on ‘conviction’, ‘leverage’, ‘lift’. Explain your findings.

top_rules_by_conviction = rules.nlargest(5, 'conviction')
top_rules_by_leverage = rules.nlargest(5, 'leverage')
top_rules_by_lift = rules.nlargest(5, 'lift')

print(top_rules_by_conviction)
print("************************")
print(top_rules_by_leverage)
print("************************")
print(top_rules_by_lift)
```

```python
# b.	Find top 2 & bottom 2 antecedent -> consequent rules based on ‘zhang’s metric’. Explain your findings.

rules['zhangs_metric'] = rules['support'] - rules['confidence']

top_rules_by_zhangs_metric = rules.nlargest(2, 'zhangs_metric')
bottom_rules_by_zhangs_metric = rules.nsmallest(2, 'zhangs_metric')

print(top_rules_by_zhangs_metric)
print("************************")
print(bottom_rules_by_zhangs_metric)
```
