```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# 1.	Firstly, replace all Missing values with relevant figures.

data = pd.read_csv('Dataset_Day10.csv')
data.head()
```

```python
colums_to_row = ['Glucose','BloodPressure','BMI','DiabetesPedigreeFunction']

data[colums_to_row] = data[colums_to_row].replace(0,float('NaN'))
```

```python
data.isna().sum()
```

```python
data.skew()
```

```python
data['Glucose'].fillna(data['Glucose'].median(),inplace=True)
data['BloodPressure'].fillna(data['BloodPressure'].mean(),inplace=True)
data['BMI'].fillna(data['BMI'].median(),inplace=True)
```

```python
data.isna().sum()
```

```python
# 2.	Then remove all existing outliers and get the final data for classification.

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# 3.	Split the data into 80% training and 20% testing data. Then, use a Decision Tree classifier algorithm with target variable as ‘Outcome’.
# a.	Print the default model performance metrics: Accuracy, Precision, Recall, F1Score
# b.	Plot a Precision & Recall vs max_leaf_nodes (consider a range of numbers) curve (both Prec and Rec on the same graph). Find the kernel type for which F1-score is the highest.
# c.	Plot a Precision & Recall vs max_depth (consider a range of numbers) curve (both Prec and Rec on the same graph). Find the kernel type for which F1-score is the highest.

X = data.drop('Outcome',axis=1)
Y = data['Outcome']

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=50)
len(X_train),len(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

dt = DecisionTreeClassifier(random_state=50)

dt = dt.fit(X_train,y_train)

y_pred = dt.predict(X_test)

print("Model Performance metrics are as below :-\n")
print("Accuracy is "+str(accuracy_score(y_test,y_pred)))
print("Precision is "+str(precision_score(y_test,y_pred)))
print("Recall is "+str(recall_score(y_test,y_pred)))
print("F1-Score is "+str(f1_score(y_test,y_pred)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

max_leaf_nodes = np.arange(2, 21)

precision_scores = []
recall_scores = []
f1_scores = []

for nodes in max_leaf_nodes:
    dt = DecisionTreeClassifier(max_leaf_nodes=nodes)
    dt.fit(X_train, y_train)
    y_pred = dt.predict(X_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    precision_scores.append(precision)
    recall_scores.append(recall)
    f1_scores.append(f1)

# Plot Precision & Recall vs max_leaf_nodes curve
plt.plot(max_leaf_nodes, precision_scores, label='Precision')
plt.plot(max_leaf_nodes, recall_scores, label='Recall')
plt.xlabel('max_leaf_nodes')
plt.ylabel('Score')
plt.title('Precision & Recall vs max_leaf_nodes')
plt.legend()
plt.show()

# Find the max_leaf_nodes value for the highest F1-score
max_f1_index = np.argmax(f1_scores)
best_max_leaf_nodes = max_leaf_nodes[max_f1_index]
print("Best max_leaf_nodes for highest F1-score:", best_max_leaf_nodes)
```

```python
max_depth_values = np.arange(1, 21)

precision_scores = []
recall_scores = []
f1_scores = []

for depth in max_depth_values:
    dt = DecisionTreeClassifier(max_depth=depth)
    dt.fit(X_train, y_train)
    y_pred = dt.predict(X_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    precision_scores.append(precision)
    recall_scores.append(recall)
    f1_scores.append(f1)

# Plot Precision & Recall vs max_depth curve
plt.plot(max_depth_values, precision_scores, label='Precision')
plt.plot(max_depth_values, recall_scores, label='Recall')
plt.xlabel('max_depth')
plt.ylabel('Score')
plt.title('Precision & Recall vs max_depth')
plt.legend()
plt.show()

# Find the max_depth value for the highest F1-score
max_f1_index = np.argmax(f1_scores)
best_max_depth = max_depth_values[max_f1_index]
print("Best max_depth for highest F1-score:", best_max_depth)
```
